```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('College_Data',index_col=0)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
def change_Grad_Rate (col):
  val = col[0]
  if val >100:
    return 100
  return val

df['Grad.Rate']=df[['Grad.Rate']].apply(change_Grad_Rate,axis =1)
```

```python
df.describe()
```

```python
sns.lmplot(x='Room.Board',y='Grad.Rate',data=df, hue='Private',palette='coolwarm',aspect=1,fit_reg=False)
```

```python
sns.lmplot(x='Outstate',y='F.Undergrad',data=df, hue='Private',palette='coolwarm',aspect=1,fit_reg=False)
```

```python
g = sns.FacetGrid(df,hue="Private",palette='coolwarm',aspect=2)
g = g.map(plt.hist,'Outstate',bins=20,alpha=0.7)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
df.columns
```

```python
kmeans.fit(df.drop('Private',axis=1))
```

```python
kmeans.cluster_centers_
```

```python
def converter(cluster):
    if cluster=='Yes':
        return 1
    else:
        return 0
df['Cluster'] = df['Private'].apply(converter)
```

```python
df.head()
```

```python
kmeans.labels_
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(df['Cluster'],kmeans.labels_))
print(classification_report(df['Cluster'],kmeans.labels_))
```
